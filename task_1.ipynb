# Task: “Why is the Tower of God Show So Popular?” Review 

```python
import matplotlib.pyplot as plt
```

```python
# Mock data for page views over time (days)
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7']
page_views = [120, 200, 150, 220, 180, 170, 210]
```

```python
# Mock data for time spent on different sections of the article
sections = ['Intro', 'Tower of God Overview', 'Korean Narrative', 'Character Design', 'Season 2 Update', 'Similar Anime']
time_spent = [15, 25, 20, 18, 10, 12]  # in percentage
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar Graph: Page Views over time
ax1.bar(days, page_views, color='skyblue')
ax1.set_title('Page Views Over Time')
ax1.set_xlabel('Days')
ax1.set_ylabel('Page Views')

# Pie Chart: Time spent on sections
ax2.pie(time_spent, labels=sections, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'gold', 'lightgreen', 'skyblue', 'plum', 'peachpuff'])
ax2.set_title('Time Spent on Article Sections')

# Show plots
plt.tight_layout()
plt.show()
```

Here are two visualizations based on the article:

1. **Bar Graph**: This shows the *page views* over time (7 days). You can see fluctuations in how many users visited the article each day.
   
2. **Pie Chart**: This visualizes the *time spent* on different sections of the article. It appears users spend the most time on the *Tower of God Overview* and *Korean Narrative*, indicating these sections are particularly engaging.

### Optimization Strategies:
1. **Interactive Elements**: Consider adding a quiz or a poll at the end of high-engagement sections (e.g., *Tower of God Overview*). This will keep users on the page for longer, as they're more likely to interact with these popular sections.
   
2. **Media Enrichment**: For the sections where time spent is lower (e.g., *Season 2 Update*), incorporate engaging media such as short trailers, character clips, or interactive timelines to maintain user interest.
